a) (4 points) Create an NLTK-corpus using PlaintextCorpusReader from the local text file created in 1c). For the tokenization of sentences please use the RegexpTokenizer with the pattern: r'[^.!?]+' . Using NLTK’s method words(), output at most 200 words of the corpus together with their absolute frequency in descending order (regarding frequency). Note: this output might contain elements like punctuation marks and stop words

```python
import nltk
from nltk.corpus import PlaintextCorpusReader
corpus_root = '' 
my_sent_tokenizer = nltk.RegexpTokenizer(r'[^.!?]+')
main_text = PlaintextCorpusReader(corpus_root, 'main_text.txt', sent_tokenizer=my_sent_tokenizer) #tokenization of sentences please use the RegexpTokenizer with the pattern: r'[^.!?]+'
main_text.fileids()
```

```python
print(main_text.sents())
```

```python
from collections import Counter
c = Counter(main_text.words()) # Using NLTK’s method words()
c.most_common(200) # output at most 200 words of the corpus together with their absolute frequency in descending order (regarding frequency)
```

b) (6 points) Filter out the following elements from the word list obtained in the previous step:
     • any stop words found in NLTK’s English stop word corpus
     • any elements that do not contain the letters A-Z and a-z (Regular expression pattern: r'[^A-Za-z]' ) 
Output this filtered list of (at most 200) words together with their absolute frequency in descending order (regarding frequency). Also, count in a defaultdict how often a word occurs that is still present after the filtering is complete

```python
import nltk
from nltk.corpus import stopwords
stopwords_en = stopwords.words('english')
print("Number of English stop words: ", len(stopwords_en))
print("English stop words: ", ", ".join(stopwords_en)) # Displaying stop words found in NLTK’s English stop word corpus
```

```python
from collections import defaultdict
import re
pruned_word_list = []
allwords = defaultdict(lambda: 0)

words = list(filter(None, [re.sub(r'[^A-Za-z]', '', word) for word in main_text.words()])) # any elements that do not contain the letters A-Z and a-z (Regular expression pattern: r'[^A-Za-z]' )
#words

for k in words:
    if k.lower() not in stopwords_en:            
        pruned_word_list.append(k.lower())
        allwords[k.lower()] += 1;
        
c = Counter(pruned_word_list)
c.most_common(200) # Output this filtered list of (at most 200) words together with their absolute frequency in descending order
```

c) (5 points) Create and output a word cloud using the defaultdict obtained in 2b)

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wc = WordCloud(background_color="white", max_words=50) # Create and output a word cloud using the defaultdict obtained
wc.generate_from_frequencies(allwords)
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```
